```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded CSV
df = pd.read_csv("netflix_titles.csv.zip")

# View first few rows
df.head()
```

```python
# Convert date_added to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Clean the data
df = df.dropna(subset=['type', 'country'])

# Plot: Movies vs TV Shows
df['type'].value_counts().plot(kind='bar', color='tomato')
plt.title("Content Type Distribution")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
# Plot: Top 10 Content-Producing Countries
df['country'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Countries by Number of Netflix Titles")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()

# Plot: Release Trend
df['release_year'] = df['release_year'].astype(int)
df['release_year'].value_counts().sort_index().plot(kind='line', color='purple')
plt.title("Netflix Releases Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.grid(True)
plt.show()
```
